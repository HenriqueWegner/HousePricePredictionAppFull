#### We can see that apparently all the NaN is type float and the others string, what I want to do now with the loop below is check if the Postal Code (which is the original column from the PPRI) is null and the Postal Town (which was the area returned from Google Maps API) is not null, we will fill Postal Code with the value from Postal Town.

```python
for x in range(len(df)):
    var1 = df.loc[x, 'Postal Code']
    var2 = df.loc[x, 'Postal Town']
    if (type(var1) == float): #if Postal Code is float means it's nan
        if (type(var2) != float): #if Postal Town is not float means it's not nan, we have a value
            df['Postal Code'][x] = df['Postal Town'][x]
```
